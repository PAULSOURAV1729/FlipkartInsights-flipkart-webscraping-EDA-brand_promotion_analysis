```python
name=[]

for i in range(1,100):
    url="https://www.flipkart.com/search?q=smartphones&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&p%5B%5D=facets.brand%255B%255D%3DSAMSUNG&p%5B%5D=facets.brand%255B%255D%3DOnePlus&p%5B%5D=facets.brand%255B%255D%3DApple&p%5B%5D=facets.price_range.from%3D20000&p%5B%5D=facets.price_range.to%3DMax&sort=price_asc&page="+str(i)
    web=requests.get(url,headers=headers)
    soup=BeautifulSoup(web.text,"lxml")
    mobiles=soup.find_all("div",class_="_2kHMtA")
    
    for j in mobiles:
        name.append(j.find("div", class_="_4rR01T").text.strip())
```

```python
len(name)
```

```python
offer_price=[]

for i in range(1,100):
    url="https://www.flipkart.com/search?q=smartphones&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&p%5B%5D=facets.brand%255B%255D%3DSAMSUNG&p%5B%5D=facets.brand%255B%255D%3DOnePlus&p%5B%5D=facets.brand%255B%255D%3DApple&p%5B%5D=facets.price_range.from%3D20000&p%5B%5D=facets.price_range.to%3DMax&sort=price_asc&page="+str(i)
    web=requests.get(url,headers=headers)
    soup=BeautifulSoup(web.text,"lxml")
    mobiles=soup.find_all("div",class_="_2kHMtA")
    
    for j in mobiles:
        element=j.find("div",class_="_30jeq3 _1_WHN1")
        if element is not None:
                text = element.text.strip()
        else:
            text = ''
        offer_price.append(text)
```

```python
actual_price=[]

for i in range(1,100):
    url="https://www.flipkart.com/search?q=smartphones&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&p%5B%5D=facets.brand%255B%255D%3DSAMSUNG&p%5B%5D=facets.brand%255B%255D%3DOnePlus&p%5B%5D=facets.brand%255B%255D%3DApple&p%5B%5D=facets.price_range.from%3D20000&p%5B%5D=facets.price_range.to%3DMax&sort=price_asc&page="+str(i)
    web=requests.get(url,headers=headers)
    soup=BeautifulSoup(web.text,"lxml")
    mobiles=soup.find_all("div",class_="_2kHMtA")
    
    for j in mobiles:
        element=j.find("div", class_="_3I9_wc _27UcVY")
        if element is not None:
                text = element.text.strip()
        else:
            text = ''
        actual_price.append(text)
```

```python
rating_review=[]

for i in range(1,100):
    url="https://www.flipkart.com/search?q=smartphones&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&p%5B%5D=facets.brand%255B%255D%3DSAMSUNG&p%5B%5D=facets.brand%255B%255D%3DOnePlus&p%5B%5D=facets.brand%255B%255D%3DApple&p%5B%5D=facets.price_range.from%3D20000&p%5B%5D=facets.price_range.to%3DMax&sort=price_asc&page="+str(i)
    web=requests.get(url,headers=headers)
    soup=BeautifulSoup(web.text,"lxml")
    mobiles=soup.find_all("div",class_="_2kHMtA")
    
    for j in mobiles:
        element=j.find("span",class_="_2_R_DZ")
        if element is not None:
                text = element.text.strip()
        else:
            text = ''
        rating_review.append(text)
```

```python
offer=[]

for i in range(1,100):
    url="https://www.flipkart.com/search?q=smartphones&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&p%5B%5D=facets.brand%255B%255D%3DSAMSUNG&p%5B%5D=facets.brand%255B%255D%3DOnePlus&p%5B%5D=facets.brand%255B%255D%3DApple&p%5B%5D=facets.price_range.from%3D20000&p%5B%5D=facets.price_range.to%3DMax&sort=price_asc&page="+str(i)
    web=requests.get(url,headers=headers)
    soup=BeautifulSoup(web.text,"lxml")
    mobiles=soup.find_all("div",class_="_2kHMtA")
    
    for j in mobiles:
        element = j.find("div", class_="_3Ay6Sb")
        if element is not None:
                text = element.text.strip()
        else:
            text = ''
        offer.append(text)
```

```python
product_links=[]

for i in range(1,100):
    url="https://www.flipkart.com/search?q=smartphones&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&p%5B%5D=facets.brand%255B%255D%3DSAMSUNG&p%5B%5D=facets.brand%255B%255D%3DOnePlus&p%5B%5D=facets.brand%255B%255D%3DApple&p%5B%5D=facets.price_range.from%3D20000&p%5B%5D=facets.price_range.to%3DMax&sort=price_asc&page="+str(i)
    web=requests.get(url,headers=headers)
    soup=BeautifulSoup(web.text,"lxml")
    mobiles=soup.find_all("div",class_="_2kHMtA")
    
    links=soup.find_all("a",class_="_1fQZEK")
    l=len(links)
    for j in range(0,l):
        pk=links[j].get("href")
        npk="https://www.flipkart.com"+str(pk)
        product_links.append(npk)
```

```python
len(product_links)
```
